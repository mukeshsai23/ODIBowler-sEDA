# Outliyer Analysis

# IQR Method

```python
# Select relevant columns for the IQR method
selected_columns = ['Matches', 'Inns', 'Balls', 'Wickets', 'Economy', 'Bowling SR', 'No of 4Wicket Haul', 'No of 5Wicket haul']
subset_df = df[selected_columns]

# Calculate the IQR for each column
Q1 = subset_df.quantile(0.25)
Q3 = subset_df.quantile(0.75)
IQR = Q3 - Q1

# Identify outliers using the IQR method
outliers = ((subset_df < (Q1 - 1.5 * IQR)) | (subset_df > (Q3 + 1.5 * IQR))).any(axis=1)

# Print the rows containing outliers
print("Rows with outliers:")
df[outliers]
```

# Hypothesis Testing for Wickets Column

```python
import pandas as pd
from scipy import stats

# Extract the 'wickets' column
wickets_data = df['Wickets']

# Define the null hypothesis (H0): Average wickets = 300
null_hypothesis_value = 300

# Perform one-sample t-test
t_statistic, p_value = stats.ttest_1samp(wickets_data, null_hypothesis_value)

# Decide on significance level (alpha)
alpha = 0.05

# Print the results
print(f'T-statistic: {t_statistic}')
print(f'P-value: {p_value}')

# Compare p-value to alpha
if p_value < alpha:
    print("Reject the null hypothesis. There is evidence to suggest a significant difference.")
else:
    print("Fail to reject the null hypothesis. There is not enough evidence to suggest a significant difference.")
```
